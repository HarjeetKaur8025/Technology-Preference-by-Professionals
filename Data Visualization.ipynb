## Demo: How to work with database


Download database file.


```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_AWWH = pd.read_sql_query(QUERY,conn)
df_AWWH.head()
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_desire = pd.read_sql_query(QUERY,conn)
df_desire
```

```python
df_top5data = df_desire.groupby('DatabaseDesireNextYear').count().sort_values(by=['Respondent'],ascending=False).reset_index()
df_top5data
```

```python
df_top5data = df_top5data.truncate(after=9)
df_top5data
```

```python
import matplotlib.pyplot as plt
import random
import matplotlib.colors as mcolors
colors = random.choices(list(mcolors.CSS4_COLORS.values()),k = 10)
sns.set()
```

```python
df_top5data.plot(kind='pie',y='Respondent',labels=df_top5data['DatabaseDesireNextYear'], colors=colors, autopct='%.2f');
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_Age_CC = pd.read_sql_query(QUERY,conn)
df_Age_CC.head()
```

```python
df_Age_CC = df_Age_CC[df_Age_CC['ConvertedComp'].notna()]
df_Age_CC.head(10)
```

```python
df45to60=df_Age_CC[df_Age_CC['Age']<61]
df45to60=df45to60[df45to60['Age']>44]
df45to60.sort_values(by='Age')
```

```python
df_med45to60=df45to60.groupby("Age").median()
df_med45to60
```

```python
sns.set_style(style='white') 
sns.lineplot(data=df_med45to60, x='Age', y="ConvertedComp")
```
